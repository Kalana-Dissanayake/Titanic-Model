```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the datasets
train_df = pd.read_csv(r'F:\ANC\Assignments\Semester 4\Artificial Intelligence,Machine Learning\train.csv')
test_df = pd.read_csv(r'F:\ANC\Assignments\Semester 4\Artificial Intelligence,Machine Learning\test.csv')

# Dropping 'Cabin' and not useful columns for the model
train_df.drop(['Cabin', 'Ticket', 'Name', 'PassengerId'], axis=1, inplace=True)
test_df.drop(['Cabin', 'Ticket', 'Name', 'PassengerId'], axis=1, inplace=True)

# Imputing missing values
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')

# One-hot encoding for categorical variables
one_hot_encoder = OneHotEncoder()

# Column transformer for different types of columns
preprocessor = ColumnTransformer(transformers=[
    ('num', num_imputer, ['Age', 'Fare']),
    ('cat', cat_imputer, ['Embarked']),
    ('ohe', one_hot_encoder, ['Sex', 'Embarked'])
])

# Apply transformations
train_df_transformed = preprocessor.fit_transform(train_df)
test_df_transformed = preprocessor.transform(test_df)

print('Data preprocessing completed.')

```

```python

# Defining the preprocessing for numerical and categorical features
numerical_features = ['Age', 'SibSp', 'Parch', 'Fare']
categorical_features = ['Embarked', 'Sex', 'Pclass']

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

# Combining preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)])

# Splitting the data into features and target variable
X = train_df.drop('Survived', axis=1)
y = train_df['Survived']

# Applying preprocessing
X_preprocessed = preprocessor.fit_transform(X)

# Splitting the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

# Training the Naive Bayes model
model = GaussianNB()
model.fit(X_train, y_train)

# Predicting on the validation set
y_pred = model.predict(X_val)

# Predicting on the training set
y_train_pred = model.predict(X_train)

# Calculating the validation accuracy
val_accuracy = accuracy_score(y_val, y_pred)

# Calculating the training accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)

#Classification Report
classification_rep = classification_report(y_val, y_pred)

# Confusion Matrix
cm = confusion_matrix(y_val, y_pred)

print('Model training and validation completed.')
print('Validation Accuracy:', val_accuracy)
print('Training Accuracy:', train_accuracy)
print('Confusion Matrix:', cm)
print(classification_rep)
```

```python

# Applying the same preprocessing pipeline used for the training data to the test data
test_df_transformed = preprocessor.transform(test_df)

# Predicting on the preprocessed test data
test_predictions = model.predict(test_df_transformed)

# Preparing the submission file with the original PassengerId
test_df_original = pd.read_csv(r'F:\ANC\Assignments\Semester 4\Artificial Intelligence,Machine Learning\test.csv')
submission_df = pd.DataFrame({'PassengerId': test_df_original['PassengerId'], 'Survived': test_predictions})
submission_df.to_csv('titanic_predictions.csv', index=False)

predictions_df = pd.DataFrame({'PassengerId': test_df_original['PassengerId'], 'Survived': test_predictions})

# Displaying the head of the predictions DataFrame
display(predictions_df.head(11))

```

```python
# Preparing data for stacked area chart
survived_counts = predictions_df['Survived'].value_counts().sort_index()

# Creating a DataFrame for plotting
plot_data = pd.DataFrame({'Did Not Survive': [survived_counts[0], 0],
                           'Survived': [0, survived_counts[1]]},
                          index=['Did Not Survive', 'Survived'])

# Plotting the stacked area chart
plt.figure(figsize=(7, 4))
plt.stackplot(plot_data.index, plot_data['Did Not Survive'], plot_data['Survived'], labels=['Did Not Survive', 'Survived'], colors=['#FF6347', '#4682B4'])
plt.title('Distribution of Predictions for Survival using Stacked Area Chart')
plt.xlabel('Status')
plt.ylabel('Count')
plt.legend(loc='upper left')
plt.show()
```

```python
survival_rate_by_sex = train_df.groupby('Sex')['Survived'].mean().reset_index()

plt.figure(figsize=(5, 5))
plt.pie(survival_rate_by_sex['Survived'], labels=survival_rate_by_sex['Sex'], autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Survival Rate by Sex')
plt.show()
```

```python
# Visualizing the survival rate by age using histograms
plt.figure(figsize=(7, 5))
sns.histplot(data=train_df, x='Age', hue='Survived', multiple='stack', bins=30, kde=True, palette='husl')
plt.title('Survival Rate by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# Visualizing the survival rate by passenger class using a countplot
plt.figure(figsize=(7, 4))
sns.countplot(data=train_df, x='Pclass', hue='Survived', palette='Set1')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# Visualizing the confusion matrix using a heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# Visualizing the accuracy
plt.figure(figsize=(6, 4))
sns.barplot(x=['Training', 'Validation'], y=[train_accuracy, val_accuracy], palette='Set1')
plt.title('Model Accuracy on Training and Validation Sets')
plt.ylabel('Accuracy')
plt.ylim(0, 1)  # Set the y-axis limit to 0-1 for accuracy
plt.show()
```
